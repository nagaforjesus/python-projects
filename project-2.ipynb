1ia.) What is recursion?

Recursion is a programming and mathematical concept where a function calls itself in order to solve a problem. In simpler terms, it's a process in which a function solves a problem by breaking it down into smaller, similar subproblems.

ib.) Write a recursive function to find the fibonacci sequence for a given number.


```python
def fibonacci(n):
    # This code will Check for the base cases
    if n < 0:
        raise ValueError("Input must be a non-negative integer.")
    elif n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        #  This will Recursively calculate Fibonacci for n-1 and n-2
        return fibonacci(n - 1) + fibonacci(n - 2)

try:  #This block begins a try-except block to handle potential exceptions that may occur.
    n = int(input("Enter a non-negative integer:" )) #this block make sure it is a non negative integer and it returns back to the value n
    if n < 0:
        print("Input must be a non-negative integer.")
    else:
        fib = fibonacci(n)
        print(f"The Fibonacci sequence value for {n} is {fib}.")
except ValueError:
    print("Invalid input. Please enter a non-negative integer.")
```

cii.) Create a child class Bus that inherits all of the variables and methods of the vehicle class. The required output should be as follows

```python

class Bus(Vehicle):
    def __init__(self, max_speed, mileage, name):
        super().__init__(max_speed, mileage)
        self.name = name

# Created an instance of the Bus class
school_bus = Bus(180, 12, "School Volvo")

# Printing the required output
print("Vehicle Name:", school_bus.name)
print("Speed:", school_bus.max_speed)
print("Mileage:", school_bus.mileage)
```

```python
class Vehicle:
    def __init__(self, max_speed, mileage):
        self.max_speed = max_speed
        self.mileage = mileage

class Bus(Vehicle):
    def __init__(self, max_speed, mileage, name):
        super().__init__(max_speed, mileage)
        self.name = name
#Defined the format to be printed out
    def seating_capacity(self, capacity=50):
        return f"The seating capacity of a {self.name} is {capacity} passengers."

school_bus = Bus(180, "12 miles per gallon", "School Volvo")

# Printing the required output
print(school_bus.seating_capacity())
```

iie.) Define a class attribute “color” with a default value yellow. Every vehicle instance should be yellow.

```python
class Vehicle:
    # Class attribute with a default value yellow
    color = "yellow"

    def __init__(self, max_speed, mileage):
        self.max_speed = max_speed
        self.mileage = mileage
        # Set the instance color to the class attribute value
        self.color = Vehicle.color

# Creating instances of the Vehicle class
car = Vehicle(180, 20000)

# Accessing and printing instance attributes and class attribute
print("Car  Color:", car.color)
```

```python
class Vehicle:
    def __init__(self, seating_capacity):
        self.seating_capacity = seating_capacity

    def calculate_fare(self):
        return self.seating_capacity * 100

class Bus(Vehicle):
    def __init__(self, seating_capacity):
        super().__init__(seating_capacity)
# Calculate the base fare using the parent class's method
    def calculate_fare(self):
        base_fare = super().calculate_fare()
        maintenance_charge = base_fare * 0.10
        total_fare = base_fare + maintenance_charge
        return total_fare

# Create instances of Bus and Vehicle
school_bus = Bus(50)
car = Vehicle(5)

# Print the required output
print(f"The seating capacity of a School bus is {school_bus.seating_capacity} passengers")
print(f"Bus fare is {school_bus.calculate_fare()}")
print(f"Car fare is {car.calculate_fare()}")
```

Vi.) Plot the histograms for main columns to show the detailed distribution of
data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loaded the data into a Pandas DataFrame
file_path = '/content/drive/MyDrive/Colab Notebooks/admissions_csv (1)/Admission_Predict_Ver1.1.csv'
data = pd.read_csv(file_path)

# Defined the main columns i want to plot histograms for
main_columns = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA', 'Research', 'Chance of Admit ']

# Created a figure with subplots for each main column
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))
fig.subplots_adjust(hspace=0.5)

# Plot histograms for the selected main columns
for i, column in enumerate(main_columns):
    row, col = i // 4, i % 4
    axes[row, col].hist(data[column], bins=20, color='skyblue', edgecolor='black')
    axes[row, col].set_title(column)
    axes[row, col].set_xlabel("Value")
    axes[row, col].set_ylabel("Frequency")

# Show the histograms
plt.show()
```

Vii.) Plot the relations between GPA and GRE. Explain the relationship
between GRA and GRE.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Loaded the data into a Pandas DataFrame
file_path = '/content/drive/MyDrive/Colab Notebooks/admissions_csv (1)/Admission_Predict_Ver1.1.csv'
data = pd.read_csv(file_path)

# Extract the CGPA and GRE Score columns
cgpa = data['CGPA']
gre_score = data['GRE Score']

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(gre_score, cgpa, c='b', marker='o', edgecolors='k', alpha=0.6)
plt.title("Relationship Between GRE Score and CGPA")
plt.xlabel("GRE Score")
plt.ylabel("CGPA")

# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(gre_score, cgpa)[0, 1]
plt.text(320, 8.5, f'Correlation: {correlation_coefficient:.2f}', fontsize=12)

# Show the plot
plt.grid(True)
plt.show()
```
